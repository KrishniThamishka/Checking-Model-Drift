## Create dummy variables

```python
def encode(df, columns, prefixes):
  df = df.copy()
  for column, prefix in zip(columns, prefixes):
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
  return df
```

```python
df = encode(
   df,
   columns = ['Side','County','State','Timezone','Airport_Code','Wind_Direction'],
   prefixes = ['SD','CO','ST','TZ','AC','WD']
)
```

3. PSI Method

```python
def calculate_psi(expected, actual, bins=10, axis=0):
    """
    Calculate the Population Stability Index (PSI) between the expected and actual distributions.
    expected: array-like, expected values of the variable
    actual: array-like, actual values of the variable
    bins: int or sequence of scalars, number of bins or bin edges
    axis: int or None, axis along which to calculate PSI. If None, calculate PSI over the flattened array.
    Returns: float, Population Stability Index (PSI)
    """
    expected_freq, expected_bins = np.histogram(expected, bins=bins)
    actual_freq, actual_bins = np.histogram(actual, bins=bins)

    # Convert frequencies to proportions
    expected_prop = expected_freq / len(expected)
    actual_prop = actual_freq / len(actual)

    # Calculate expected proportion
    expected_prop[expected_prop == 0] = 0.001
    expected_prop[expected_prop == 1] = 0.999

    # Calculate actual proportion
    actual_prop[actual_prop == 0] = 0.001
    actual_prop[actual_prop == 1] = 0.999

    # Calculate PSI
    psi = np.sum((expected_prop - actual_prop) * np.log(expected_prop / actual_prop))

    return psi
```
